```python
import pandas as pd
from sqlalchemy import create_engine

#ADD SERVER INFORMATION HERE
username = 'postgres'
password = 'databasepassword'
db_name = 'movies_db'
```

```python
#READING IN CSV WITH MEDIA STREAMING ON NETFLIX
netflix_file = "Resources/netflix_titles.csv"
netflix_full_df = pd.read_csv(netflix_file)
netflix_full_df.head()
```

```python
#CLEANING DATA WITH ONLY THE COLUMNS NEEDED
netflix_df = netflix_full_df[[
    'show_id',
    'title',
    'type',
    'director',
    'country',
    'release_year',
    'rating',
    'duration',
    'listed_in',
    'description']].copy()

netflix_df.head()
```

```python
#READING IN CSV WITH IMDB MEDIA INFORMATION
imdb_file = "Resources/IMDB_movies.csv"
imdb_full_df = pd.read_csv(imdb_file)
imdb_full_df.head()
```

```python
##CLEANING DATA WITH ONLY THE COLUMNS NEEDED
imdb_df = imdb_full_df[[
    'imdb_title_id',
    'title',
    'year',
    'genre',
    'budget',
    'usa_gross_income',
    'worlwide_gross_income',
    'metascore',
    'reviews_from_users',
    'reviews_from_critics']].copy()

imdb_df.head()
```

```python
#USING SQLALCHEMY TO CONNECT TO POSTGRESQL DATABASE
db_info = f'postgresql://{username}:{password}@localhost:5432/{db_name}'
engine = create_engine(db_info)
```

```python
#VERIFY THAT TABLES EXIST IN THE DATABASE
#TABLES NAMED NETFLIX_DATA AND IMDB_DATA SHOULD APPEAR BELOW
engine.table_names()
```

```python
#LOADING NETFLIX DATAFRAME INTO NETFLIX_DATA TABLE IN DATABASE
netflix_df.to_sql(name='netflix_data', con=engine, if_exists='append', index=False)
```

```python
#LOADING IMDB DATAFRAME INTO IMDB_DATA TABLE IN DATABASE
imdb_df.to_sql(name='imdb_data', con=engine, if_exists='append', index=False)
```
